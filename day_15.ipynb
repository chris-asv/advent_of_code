```python
score = 0
for i in range(0,101):
    for j in range(0,101-i):
        for k in range(0,101-i-j):
            l = 100-i-j-k
            teaspoons = np.array([i,j,k,l])
            # print(teaspoons.sum())
            values = np.dot(matrix_coef, teaspoons)
            if (values <= 0).sum() >= 1:
                pass
            else:
                score = max([score, np.prod(values)])
```
